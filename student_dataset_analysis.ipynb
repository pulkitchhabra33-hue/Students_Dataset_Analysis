```python
#Some Good Plots
sns.barplot(x= 'Gender', y= 'GPA', data= df, color= 'skyblue')
plt.title('Gender vs GPA of Students')
plt.show()
```

```python
sns.barplot(x= 'Tutoring', y= 'GPA', data= df, color= 'orange')
plt.title('Tutoring vs GPA of Students')
plt.show()
```

```python
sns.barplot(x= 'ParentalSupport', y= 'GPA', data= df, color= 'orange')
plt.title('Parental Support vs GPA of Students')
plt.show()

# Very Low parental support --> 0, 
# Low parental support --> 1, 
# Medium support --> 2,	
# High support --> 3,
# Very High parental support --> 4
```

```python
sns.heatmap(df[['Age','StudyTimeWeekly','Absences','GPA']].corr(), annot=True)
plt.show()
```

```python
sns.boxplot(x= 'Extracurricular', y= 'Absences', data= df, color= 'red')
plt.title('Extracurricular of Students vs Absences of Students')
plt.show()
```

```python
(df['GPA']>3).mean()
```

```python
sns.regplot(x= 'StudyTimeWeekly', y= 'GPA', data= df.head(50))
plt.title('Weekly Time Study of Students vs their GPA')
plt.show()
```

```python
sns.barplot(x= 'Absences', y= 'GPA', data= df, color= 'pink')
plt.title('Absences of the Students vs their GPA')
plt.xticks(rotation= 90)
plt.show()
```

```python
sns.barplot(x= 'ParentalEducation', y= 'GPA', data= df, color= 'orange')
plt.title('Parental Education Data and the Students GPA')
plt.show()

# No formal education / Primary school --> 0,
# Middle school --> 1,
# High school --> 2,
# Bachelorâ€™s degree --> 3,
# Masterâ€™s / Professional degree --> 4
```

```python
# Average GPA Based on Parental Education Level
df.groupby('ParentalEducation')['GPA'].mean()
```

```python
sns.countplot(x= 'ParentalEducation', data= df)
plt.title("Parental Education Distribution")
plt.show()
```
